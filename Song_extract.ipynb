```python
#Importing video and converting it to mp3
import os
from pytube import YouTube
from pydub import AudioSegment

# Download the YouTube video
video_url = "https://www.youtube.com/watch?v=G_NKYDSI574"
yt = YouTube(video_url)
video = yt.streams.first().download()

# Convert the video to an mp3 file
audio = AudioSegment.from_file(video, "mp4")
audio.export("song.mp3", format="mp3")

# Delete the original video file
os.remove(video)
```

```python
from spleeter.separator import Separator

# Define the audio file to be separated
audio_file = 'song.mp3'
separator = Separator('spleeter:5stems')


# Perform the separation
separator.separate_to_file(audio_file, 'output/')
```
