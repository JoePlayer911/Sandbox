```python
nums = [2,7,11,15]; target = 9
dd = dict()
for i, m in enumerate(nums):
  dd[m] = i
for i, m in enumerate(nums):
  n = target - m
  if n in dd:
    j = dd[n]
    if j <= i:
      continue
    print(i, j)
```
