```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return "Target Not Found"
arr = [1, 3, 5, 7, 9, 11, 13]
target = 7

binary_search(arr, target)
```

```python
def find_duplicate(A):
    tortoise = A[0]
    hare = A[0]
    
    # detect the cycle
    while True:
        tortoise = A[tortoise]
        hare = A[A[hare]]
        if tortoise == hare:
            break
    
    # find the starting point of the cycle
    tortoise = A[0]
    while tortoise != hare:
        tortoise = A[tortoise]
        hare = A[hare]
    
    # return the repeated integer
    return tortoise

Array = [1,2,3,4,5,6,7,8,3,9]

find_duplicate(Array)
```

```python
def find_duplicate_sorting(A):
    n = len(A)
    A.sort()  # use quicksort or merge sort for O(nlogn) time complexity

    for i in range(1, n):
        if A[i] == A[i-1]:
            return A[i]

    return None  # no duplicate found
find_duplicate_sorting(Array)
```
