```python
# Take last 4 digits from a set of numbers
a = int(input("Insert number: "))
d4 = (a%10000)//1000
d3 = (a%1000)//100
d2 = (a%100)//10
d1 = (a%10)
print(f"{d4} {d3} {d2} {d1}")
```
