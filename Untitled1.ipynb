```python
import os
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from collections import Counter

# 下载NLTK punkt分词器（如有需要）
nltk.download('punkt')

# 要搜索的目录中的 .txt 文件列表
dir_path = "/path/to/directory"

# 要转换为小写字母的文件名列表
files_to_convert = []

# 循环遍历目录中的文件，并将 .txt 文件添加到列表中
for filename in os.listdir(dir_path):
    if filename.endswith(".txt"):
        files_to_convert.append(os.path.join(dir_path, filename))

# 创建一个包含所有小写字母单词的列表
lowercase_words = []
lowercase_words_without_punct = []
lowercase_words_without_punct_stopwords = []
lowercase_words_stemmed = []

# 初始化 NLTK Porter 词干提取器和停用词
stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))

# 循环遍历列表中的每个文件并处理其内容
for filename in files_to_convert:
    with open(filename, 'r') as f:
        contents = f.read()
        # 使用 NLTK punkt 分词器将内容标记为单词
        words = nltk.word_tokenize(contents.lower())
        # 使用字符串模块从每个单词中删除标点符号
        words_without_punct = [word.translate(str.maketrans('', '', string.punctuation)) for word in words]
        # 使用 NLTK 停用词移除停用词
        words_without_punct_stopwords = [word for word in words_without_punct if word not in stop_words]
        # 使用 NLTK porter 词干提取器提取词干
        words_stemmed = [stemmer.stem(word) for word in words_without_punct_stopwords]
        lowercase_words.extend(words)
        lowercase_words_without_punct.extend(words_without_punct)
        lowercase_words_without_punct_stopwords.extend(words_without_punct_stopwords)
        lowercase_words_stemmed.extend(words_stemmed)

# 打印包含所有小写字母单词的列表
print(lowercase_words)

# 打印没有标点符号的小写字母单词列表
print(lowercase_words_without_punct)

# 打印没有标点符号和停用词的小写字母单词列表
print(lowercase_words_without_punct_stopwords)

# 打印不带标点符号、停用词和词干的小写字母单词列表
print(lowercase_words_stemmed)
```
