```python
#Web Videos
import urllib.request
from bs4 import BeautifulSoup
import os
import re

# Define the URL of the website to be scraped
url = input("Insert URL: ")

# Make an HTTP request and get the HTML content
html_content = urllib.request.urlopen(url).read()

# Use BeautifulSoup to parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Find all the <a> tags on the page
links = soup.find_all('a')

# Iterate over the links and search for video files
for link in links:
    if link.has_attr('href'):
        link_url = link['href']
        if re.search(r'(\.mp4|\.mkv|\.avi|\.wmv\.m4s|)', link_url):
            local_filename = os.path.join(os.getcwd(), os.path.basename(link_url))
            urllib.request.urlretrieve(link_url, local_filename)
            print(f"{local_filename} downloaded successfully")
```

```python
import tweepy
import requests

# Add your Twitter API credentials here
consumer_key = "your_consumer_key"
consumer_secret = "your_consumer_secret"
access_token = "your_access_token"
access_token_secret = "your_access_token_secret"

# Authenticate with the Twitter API
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

# Create a Tweepy API object
api = tweepy.API(auth)

# Get the tweet ID from the tweet URL
tweet_url = "https://twitter.com/Twitter/status/1394671190637405697"
tweet_id = tweet_url.split("/")[-1]

# Get the tweet object
tweet = api.get_status(tweet_id, tweet_mode="extended")

# Get the video URL from the tweet object
media = tweet.entities.get("media", [])
if len(media) == 0:
    print("No video found in tweet.")
    exit()
video_url = media[0]["video_info"]["variants"][0]["url"]

# Download the video file
response = requests.get(video_url, stream=True)
with open("tweet.mp4", "wb") as f:
    for chunk in response.iter_content(chunk_size=1024):
        if chunk:
            f.write(chunk)
```
