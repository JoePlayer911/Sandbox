```python
import tensorflow as tf

# Load the model
model = tf.keras.models.load_model("model.h5")

# Read the low-resolution image
lr_image = tf.image.decode_jpeg(tf.io.read_file("memories.jpg"))

# Perform super-resolution
sr_image = model(tf.expand_dims(lr_image, 0))[0]

# Convert the image back to uint8 and save
sr_image = tf.image.convert_image_dtype(sr_image, tf.uint8)
tf.io.write_file("sr_image.jpg", tf.image.encode_jpeg(sr_image))
```
