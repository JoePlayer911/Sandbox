```python
import scrapy
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
class focustaiwanSpider(scrapy.Spider):
    name = 'focustaiwan'
    allowed_domains = ['https://focustaiwan.tw/']
    start_urls = ['https://focustaiwan.tw/']

    def parse(self, response):
        pass
newsTitles = BeautifulSoup.findAll('span', {'class':'h2t'})
linkset = set()
f = open("ft_crawler.json", 'w')
for link in BeautifulSoup.find_all('a', {'href':re.compile('https://focustaiwan.tw')}):
    if 'href' in link.attrs:
      curlink = link.attrs['href']
      if curlink not in linkset:
        linkset.add(curlink)
        f.write(curlink)
        f.write("\n")
f.close()
```
