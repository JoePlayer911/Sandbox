```python
#Importing video and converting it to mp3
import os
from pytube import YouTube
from pydub import AudioSegment

# Download the YouTube video
video_url = "https://youtu.be/V6s8gTz5qv0"
yt = YouTube(video_url)
video = yt.streams.first().download()

# Convert the video to an mp3 file
audio = AudioSegment.from_file(video, "mp4")
audio.export("song.mp3", format="mp3")

# Delete the original video file
os.remove(video)
```

```python
import librosa

def detect_bpm(filepath):
    # Load the audio file
    y, sr = librosa.load(filepath)
    # Detect the tempo
    tempo, beat_frames = librosa.beat.beat_track(y=y, sr=sr)
    # Print the tempo
    print("BPM: {:.2f}".format(tempo))

# Example usage
detect_bpm("song.mp3")
```
