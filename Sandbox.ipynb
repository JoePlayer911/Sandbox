```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if not self.is_empty():
            return self.stack.pop()

    def peek(self):
        if not self.is_empty():
            return self.stack[-1]

    def is_empty(self):
        return len(self.stack) == 0

    def __len__(self):
        return len(self.stack)
```

```python
class Queue:
    def __init__(self):
        self.items = []
        
    def enqueue(self, item):
        self.items.insert(0, item)
        
    def dequeue(self):
        if not self.is_empty():
            return self.items.pop()
        
    def is_empty(self):
        return len(self.items) == 0
    
    def __len__(self):
        return len(self.items)
```

```python
#Stack class
s = Stack()
s.push(1)
s.push(2)
s.push(3)
print(len(s))
print(s.pop())
print(s.pop())
print(s.peek())
```

```python
#Queue class
q = Queue()
q.enqueue('apple')
q.enqueue('banana')
q.enqueue('cherry')
print(len(q))
print(q.dequeue())
print(q.dequeue())
print(q.is_empty())
```

```python
def empty_stack(stack):
    if not stack:
        return
    else:
        stack.pop()
        empty_stack(stack)
```

```python
def search_and_restore(S, Q, x):
    found = False

    while len(S) != 0:
        Q.append(S.pop())

    while len(Q) != 0:
        elem = Q.pop(0)
        if elem == x:
            found = True
        S.append(elem)

    return found
```

```python
def permute(nums):
    if len(nums) == 0:
        return []
    if len(nums) == 1:
        return [nums]
    result = []
    for i in range(len(nums)):
        num = nums[i]
        remaining = nums[:i] + nums[i+1:]
        for p in permute(remaining):
            result.append([num] + p)
    return result

n = 3
nums = [i+1 for i in range(n)]
permutations = permute(nums)
print(permutations)
```

```python
def generate_subsets(T):
    Q = [[]]
    S = []
    
    while Q:
        subset = Q.pop(0)
        S.append(subset)
        
        for e in T:
            new_subset = subset + [e]
            Q.append(new_subset)
    
    subsets = [S.pop() for _ in range(len(S))]
    return subsets
```
