```python
#@title Encoder
###ENCODER###

import pandas as pd
import random
import string

alphabets = list(string.ascii_uppercase)
random.shuffle(alphabets)
encoder = ''.join(alphabets)
message = input("Input your message: ")
message = message.upper()
safemessage = [x for x in message]

replace_chars = {'A':alphabets[0], 'B':alphabets[1], 'C':alphabets[2],
                 'D':alphabets[3], 'E':alphabets[4], 'F':alphabets[5],
                 'G':alphabets[6], 'H':alphabets[7], 'I':alphabets[8],
                 'J':alphabets[9], 'K':alphabets[10], 'L':alphabets[11],
                 'M':alphabets[12], 'N':alphabets[13], 'O':alphabets[14], 
                 'P':alphabets[15], 'Q':alphabets[16], 'R':alphabets[17], 
                 'S':alphabets[18], 'T':alphabets[19], 'U':alphabets[20], 
                 'V':alphabets[21], 'W':alphabets[22], 'X':alphabets[23], 
                 'Y':alphabets[24], 'Z':alphabets[25], ' ':'0', ',':'9', '.':'8'}
C = (pd.Series(safemessage)).map(replace_chars) #convert the list to a pandas series temporarily before mapping
done = ''.join(list(C))
print(encoder + done)
```

```python
#@title Decoder
###DECODER###
import pandas as pd
s = input("Please input the encoded message: ")
z = s[26::]
savedmessage = [x for x in z]
decode = []

while s:
    decode.append(s[0:1:1])
    s = s[1:26:]

place_chars = {decode[0]:'A', decode[1]:'B', decode[2]:'C',
                 decode[3]:'D', decode[4]:'E', decode[5]:'F',
                 decode[6]:'G', decode[7]:'H', decode[8]:'I',
                 decode[9]:'J', decode[10]:'K', decode[11]:'L',
                 decode[12]:'M', decode[13]:'N', decode[14]:'O', 
                 decode[15]:'P', decode[16]:'Q', decode[17]:'R', 
                 decode[18]:'S', decode[19]:'T', decode[20]:'U', 
                 decode[21]:'V', decode[22]:'W', decode[23]:'X', 
                 decode[24]:'Y', decode[25]:'Z', '0':' ', '9':',', '8':'.'}
D = (pd.Series(savedmessage)).map(place_chars) #convert the list to a pandas series temporarily before mapping
decoded = ''.join(list(D))
print(decoded.lower())
```
