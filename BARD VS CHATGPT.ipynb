```python
class Node:
  data = None
  next = None

  def __init__(self, data):
    self.data = data
    self.next = None


class LinkedList:
  head = None
  tail = None

  def __init__(self):
    self.head = None
    self.tail = None

  def add(self, data):
    new_node = Node(data)

    if self.head is None:
      self.head = new_node
      self.tail = new_node
    else:
      self.tail.next = new_node
      self.tail = new_node

  def remove(self, data):
    if self.head is None:
      return

    if self.head.data == data:
      self.head = self.head.next
    else:
      current = self.head

      while current.next is not None and current.next.data != data:
        current = current.next

      if current.next is not None:
        current.next = current.next.next

  def print(self):
    current = self.head

    while current is not None:
      print(current.data, end=" ")
      current = current.next

    print()
```

```python
from itertools import cycle, tee


class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class CircularLinkedList:
    def __init__(self, data):
        self.head = Node(data)
        self.tail = self.head
        self.head.next = self.tail

    def append(self, data):
        new_node = Node(data)
        self.tail.next = new_node
        self.tail = new_node

    def split_circular_linked_list_in_half(self):
        """Splits the circular linked list into two halves of equal size.

        Returns:
            A tuple of two circular linked lists.
        """
        first_half_iter = iter(self.head)
        second_half_iter = tee(first_half_iter)[1]
        next(second_half_iter)
        return first_half_iter, second_half_iter

    def display(self):
        for node in self.head:
            print(node.data)


# Create a circular linked list with even number of nodes
my_list = CircularLinkedList(10)
my_list.append(20)
my_list.append(30)
my_list.append(40)
my_list.append(50)
my_list.append(60)

# Split the circular linked list into two halves
first_half_head, second_half_head = my_list.split_circular_linked_list_in_half()

# Display the first half
print("First Half:")
for node in first_half_head:
    print(node.data)

# Display the second half
print("Second Half:")
for node in second_half_head:
    print(node.data)
```

```python
class CircularLinkedList:
  """A circular linked list.

  Attributes:
    head: The head of the list.
  """

  def __init__(self):
    self.head = None

  def add(self, data):
    """Adds a node to the end of the list.

    Args:
      data: The data to add to the list.
    """
    new_node = Node(data)
    if self.head is None:
      self.head = new_node
      new_node.next = new_node
    else:
      current = self.head
      while current.next != self.head:
        current = current.next
      current.next = new_node
      new_node.next = self.head

  def reverse(self):
    """Reverses the order of the list.

    Returns:
      A new circular linked list with the order reversed.
    """
    if self.head is None:
      return self

    previous = None
    current = self.head

    while current is not None and current.next is not None:
      next = current.next
      current.next = previous
      previous = current
      current = next

    self.head = previous
    return self

  def __iter__(self):
    """Iterates over the list.

    Returns:
      An iterator over the list.
    """
    current = self.head
    while current is not None:
      yield current.data
      current = current.next
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            new_node.next = self.head
        else:
            current = self.head
            while current.next != self.head:
                current = current.next
            current.next = new_node
            new_node.next = self.head

    def split_in_half(self):
        if self.head is None:
            return None, None

        slow = self.head
        fast = self.head.next.next

        while fast != self.head and fast.next != self.head:
            slow = slow.next
            fast = fast.next.next

        second_half_head = slow.next
        slow.next = self.head

        current = second_half_head
        while current.next != self.head:
            current = current.next
        current.next = second_half_head

        return self.head, second_half_head

    def display(self):
        current = self.head
        while True:
            print(current.data)
            current = current.next
            if current == self.head:
                break
```

```python
def removeRange(A, i, j):
    if i > j or A is empty:
        return A

    current = A.head
    previous = None

    # Traverse to the i-th node
    for _ in range(i):
        previous = current
        current = current.next

        if current is None:
            return A

    # Remove nodes in the range [i, j]
    if previous is None:
        A.head = current.next
    else:
        previous.next = current.next

    # Traverse to the j-th node
    for _ in range(j - i):
        current = current.next

        if current is None:
            return A

    # Remove nodes in the range [i, j]
    if previous is None:
        A.head = current.next
    else:
        previous.next = current.next

    return A
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
        self.previous = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, value):
        new_node = Node(value)

        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next is not None:
                current = current.next
            current.next = new_node
            new_node.previous = current

    def display(self):
        elements = []
        current = self.head
        while current is not None:
            elements.append(str(current.value))
            current = current.next
        print(" <-> ".join(elements))
```

```python
def reverse(A):
    if A.head is None or A.head.next is None:
        return A

    current = A.head
    while current is not None:
        next_node = current.next
        current.next = current.previous
        current.previous = next_node
        if next_node is None:
            A.head = current
        current = next_node

    return A
```
