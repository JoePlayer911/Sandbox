```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bts
import re
html = urlopen("https://focustaiwan.tw/")
bs = bts(html.read(), 'html.parser')
newsTitles = bs.findAll('span', {'class':'h2t'})
linkset = set()
tab = open("ft_crawler.json", 'w')
for link in bs.find_all('a', {'href':re.compile('^https://focustaiwan\.tw')}):
    if 'href' in link.attrs:
      curlink = link.attrs['href']
      if curlink not in linkset:
        linkset.add(curlink)
        tab.write(curlink)
        tab.write("\n")
tab.close()
```
