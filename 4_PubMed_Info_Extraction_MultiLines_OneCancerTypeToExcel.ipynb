```python
CancerType = "Lung"
Inputfile ="pubmed-lungcancer-set.txt"
OutputFile = "pubmed-lungcancer-set_Output.xlsx"

#CancerType = "Breast"
#Inputfile="pubmed-breastcanc-set.txt"
#OutputFile = "pubmed-Breastcanc-set_Output.xlsx"

#CancerType = "Colorectal"
#Inputfile="pubmed-colorectal-set.txt"
#OutputFile = "pubmed-Colorectal-set_Output.xlsx"

#CancerType = "Prostate"
#Inputfile ="pubmed-Top4_Prostateca-set.txt"
#OutputFile = "pubmed-Top4-Prostateca-set_Output.xlsx"

#CancerType = "TestType"
#Inputfile ="PubMed_Test.txt"
#OutputFile = "pubmed_Test_MultiLine_Output.xlsx"

InputDir = "/content/"
OutputDir = "/content/Output_MultiLines/"

InputFileLocation = InputDir + Inputfile
OutputLoction = OutputDir+OutputFile
```

```python

#with open (file, 'r',encoding="utf-8") as file_inpu
with open (InputFileLocation, 'r',encoding="utf-8") as file_input:
    Lines = file_input.readlines()
    #print (Lines)
cnt = 0
LineIndex = 0
PMID_LineIndex = []
for OneLine in Lines:
    LineIndex = LineIndex + 1
    if (OneLine.startswith('PMID-')):
        cnt = cnt +1
        PMID_LineIndex.append(LineIndex)
PMID_LineIndex.append(LineIndex)
print(cnt)    
print(PMID_LineIndex)
```

```python
PUMID_List = []
TI_List = []
AB_List = []
with open (InputFileLocation, 'r',encoding="utf-8") as file_input:
    for i in range(0,cnt):
        StartLineIndex = PMID_LineIndex[i]
        EndLineIndex = PMID_LineIndex[i+1]
        OnePUMID = []
        for j in range(StartLineIndex,EndLineIndex):
            OneLine = file_input.readline()
            #OneLine = OneLine.strip()
            #print(OneLine)
            OnePUMID.append(OneLine)
        #print(len(OnePUMID))
        #print(OnePUMID)
        #for k in range(len(OnePUMID)):
        OnePubMed_LineCnt = len(OnePUMID)
        #print(OnePubMed_LineCnt)
        OneTI_Info = []
        OneAB_Info = []
        k=0
        while (k < OnePubMed_LineCnt):
            #print(str(k)+'=>'+OnePUMID[k])
            if (OnePUMID[k].startswith('PMID-')):
                OnePUMID_Info = OnePUMID[k].split('- ')
                #print(str(k)+' PMID:'+OnePUMID_Info[1])
            #if (OnePUMID[k].startswith('TI  -')): # 
            if (OnePUMID[k].startswith('TI  -') or OnePUMID[k].startswith('Ti  -')):
                OneTI_Info = OnePUMID[k].split('- ')
                OneTI_Info[1] = OneTI_Info[1].strip()
                while (OnePUMID[k+1].startswith('   ')):
                    Temp = OnePUMID[k+1].split('    ')
                    #print(OnePUMID[k+1])
                    OneTI_Info[1] = OneTI_Info[1]+' '+Temp[1].strip()
                    k += 1
                #print(str(k)+' TI:'+OneTI_Info[1])
            if (OnePUMID[k].startswith('AB  -')):
                OneAB_Info = OnePUMID[k].split('- ')
                OneAB_Info[1] = OneAB_Info[1].strip()
                while (OnePUMID[k+1].startswith('   ')):
                    Temp = OnePUMID[k+1].split('    ')
                    #print(OnePUMID[k+1])
                    OneAB_Info[1] = OneAB_Info[1]+' '+Temp[1].strip()
                    k += 1
               # print(str(k)+' AB:'+OneAB_Info[1])
            k += 1
        PUMID_List.append(OnePUMID_Info[1].strip())
        TI_List.append(OneTI_Info[1].strip())
        AB_List.append(OneAB_Info[1].strip())
#PUMID_List
#TI_List
AB_List


PubMed_Data = [[PUMID_List[i],TI_List[i],AB_List[i]] for i in range(cnt)]
#PubMed_Data

import pandas as pd

PubMed = pd.DataFrame(PubMed_Data, columns=['PUMID','TI','AB'])
#PubMed

PubMed['Type'] = CancerType

PubMed.to_excel(OutputLoction)
```
