```python
import scrapy
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
html = urlopen("https://focustaiwan.tw/")
bs = BeautifulSoup(html.read(), 'html.parser')
class focustaiwanSpider(scrapy.Spider):
    name = 'focustaiwan'
    allowed_domains = ['https://focustaiwan.tw/']
    start_urls = ['https://focustaiwan.tw/']

    def parse(self, response):
        pass
newsTitles = bs.findAll('span',{'class':'h2t'})
linkset = set()
f = open("ft_crawler.json", 'w')
for link in bs.find_all('a', {'href':re.compile('https://focustaiwan.tw')}):
  if 'href' in link.attrs:
    clink = link.attrs['href']
    if clink not in linkset:
      linkset.add(clink)
      f.write(f"{clink}\n")
f.close()
```
