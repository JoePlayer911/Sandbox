```python
#Q1 Unit conversion: height and weight
h = int(input('Insert height in cm: '))
w = int(input('Insert weight in kg: '))
i = h * 2.54
l = w * 0.454
iz = "{:,.1f}".format(i)
lz = "{:,.1f}".format(l)
print(f'{h} cm = {iz} inch')
print(f'{w} kg = {lz} pounds')
print('')
```

```python
#Q3 Unit conversion: liters to gallons
l = float(input('Insert liters: '))
g = l * 0.26418
gal = "{:,.1f}".format(g)
print(f'{l} liters = {gal} galons')
print('')
```

```python
#Q5 Calculate the sum, product, difference, quotient and remainder
a = int(input('Input first integer: '))
b = int(input('Input second integer: '))

s = a + b
p = a * b
d = a - b
z = float("{:,.0f}".format(a / b))
r = a % b
if z > (a / b):
  q = "{:,.0f}".format(z-1)
else:
  q = "{:,.0f}".format(z)

print(f'{a}+{b}={s}')
print(f'{a}*{b}={p}')
print(f'{a}-{b}={d}')
print(f'{a}/{b}={q}...{r}')
print('')

```

```python
#Q6 Interval formula
x = int(input('Insert integer: '))
y=0
if x < -1:
  y = 3 * (x**2)
if -1 <= x <= 1:
  y = (x**3) + 3 * x - 4
if x > 1:
  y = 3 * x + 6
print(y)
print('')
```

```python
#Q7 Calculate electricity bill
deg = int(input('Insert degree of usage: '))
bs = float()
bns = float()
if deg <= 120:
 bs = 2.10
 bns = 2.10
elif 121 <= deg <= 330:
 bs = 3.02
 bns = 2.68
elif 331 <= deg <= 500:
 bs = 4.39
 bns = 3.61
elif 501 <= deg <= 700:
 bs = 4.97
 bns = 4.01
elif deg >= 701:
 bs = 5.63
 bns = 4.50
bsz = bs*deg
bnsz = bns*deg
sum = "{:,.2f}".format(bsz)
nonsum = "{:,.2f}".format(bnsz)
print('Summer electricity bill:', sum)
print('Non summer electricity bill:', nonsum)
print('')
```

```python
#Q8 Triangle classification
print('Input three side lengths a, b, c (and a≤b≤c)')
a = int(input('First side: '))
b = int(input('Second side: '))
c = int(input('Third side: '))
x = int(a**2 + b**2)
y = int(c**2)

if x < y:
 print('obtuse triangle')
 print('')
if x > y:
 print('acute triangle')
 print('')
if x is y:
 print('right triangle')
 print('')
```
