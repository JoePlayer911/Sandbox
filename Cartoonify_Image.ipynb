```python
from PIL import Image, ImageFilter

# Open the image
im = Image.open("memories.jpg")

# Apply a "cartoon" filter
im = im.filter(ImageFilter.EDGE_ENHANCE_MORE)

# Convert the image to grayscale
im = im.convert("RGB")

# Apply a "posterize" filter to reduce the number of colors
im = im.convert("RGB", palette=Image.ADAPTIVE, colors=256)

# Save the cartoonified image
im.save("cartoon.png")
```
